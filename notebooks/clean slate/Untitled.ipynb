# PG4

```julia
nms = [(1, 1), (1, 2), (2, 1), (2, 2), (1, 3), (3, 1), (2, 3), (3, 2)]
gs = []
probs = []
for (n, m) in nms
    @time gs_ = get_bicoloured_graphs(n, m)
    @show length(gs_)
    probs_ = prob.(gs_, n, m)
    append!(gs, gs_)
    append!(probs, probs_)
end
```

```julia
scatter(sort(probs))
```

# PG3

```julia
for n in 1:3
    for m in 1:3
        if n == m == 3
            break
        end
        @show n, m
        @time gs = get_bicoloured_graphs(n, m);
        @time gs = filter(condition1, gs)
        @show length(gs);
        gs = gs[sortperm(ne.(gs))];

        many_gvs("../../data/si_geoms/n$(n)_m$(m).pdf", gs, n, m)
    end
end
```

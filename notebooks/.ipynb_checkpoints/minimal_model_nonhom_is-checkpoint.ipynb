### 1D spatial

```julia
function add_peak_1d!(data, dx, center_i, width, mag=1.; mirrors=1)
    ssize = length(data)
    centers = [center_i]
    for i in 1:mirrors
        push!(centers, center_i - i*ssize)
        push!(centers, center_i + i*ssize)
    end

    sigma = width / 2.355
    a = mag / (sigma * sqrt(2 * pi))
    b = 2 * sigma^2
    
    for i in 1:ssize
        for c in centers
            dist = abs(i - c) * dx
            peak_val = a * exp(-dist^2 / b)
            data[i] += peak_val
        end
    end
    data
end

function add_peaks_1d!(data, dx, num_peaks, width, mag=1.; kwargs...)
    ssize = length(data)
    for _ in 1:num_peaks
        center_i = rand(1:ssize)
        add_peak_1d!(data, dx, center_i, width, mag; kwargs...)
    end
    data
end
```

```julia
sys_size = (1000,)
dx = 0.01

# initiate with glucose at max, and no N or R
u0 = expand_u0_to_size(sys_size, [0., mmps.K, 0.]);
# but add a bunch of peaks of N
add_peaks_1d!((@view u0[1,:]), dx, 10, 0.2, 1.; mirrors=1);
# add_peaks_1d!((@view u0[1,:]), dx, 10, 0.2, 1.);
```

```julia
spatial_params = change_sasmmicrm_params(smmicrm_params;
    space=make_cartesianspace_smart(1;
        dx,
        bcs=Periodic()
    ),
    usenthreads=nthreads()
)
sp = make_smmicrm_problem(spatial_params, u0, 1000000)
```

```julia
sps = solve(sp, QNDF())
print_spatial_solution_stats(sps)
plot_1dsmmicrm_sol_snap(sps, -1)
```

```julia
f = plot_1dsmmicrm_sol_interactive(sps)
display(GLMakie.Screen(), f)
```

### 2D spatial

```julia
function add_peak_2d!(data, dx, center, width, mag=1.; mirrors=1)
    sx, sy = size(data)
    centers = [center]
    for i in 1:mirrors
        push!(centers, (center[1] - i*sx, center[2]))
        push!(centers, (center[1] + i*sx, center[2]))
        push!(centers, (center[1], center[2] - i*sy))
        push!(centers, (center[1], center[2] + i*sy))
    end

    sigma = width / 2.355
    a = mag / (sigma * sqrt(2 * pi))
    b = 2 * sigma^2
    
    for i in 1:sx
        for j in 1:sy
            for (cx, cy) in centers
                dist = sqrt((abs(i - cx) * dx[1])^2 + (abs(j - cy) * dx[2])^2)
                peak_val = a * exp(-dist^2 / b)
                data[i,j] += peak_val
            end
        end
    end
    data
end

function add_peaks_2d!(data, dx, num_peaks, width, mag=1.)
    sx, sy = size(data)
    for _ in 1:num_peaks
        cx = rand(1:sx)
        cy = rand(1:sy)
        add_peak_2d!(data, dx, (cx, cy), width, mag)
    end
    data
end
```

```julia
sys_size = (500, 500)
dx = [0.01, 0.01]

# initiate with glucose at max, and no N or R
u0 = expand_u0_to_size(sys_size, [0., mmps.K, 0.]);
# but add a bunch of peaks of N
add_peaks_2d!((@view u0[1,:,:]), dx, 30, 0.3, 2.);
```

```julia
spatial_params = change_sasmmicrm_params(smmicrm_params;
    space=make_cartesianspace_smart(2; dx, bcs=Periodic()),
    usenthreads=nthreads()
)
sp = make_smmicrm_problem(spatial_params, u0, 1000000000000);
size(sp.u0)
```

```julia
sps = solve(sp, QNDF())
print_spatial_solution_stats(sps)
plot_2dsmmicrm_sol_snap_heatmap(sps, -1)
```

```julia
f = plot_2dsmmicrm_sol_interactive_heatmap(sps)
display(GLMakie.Screen(), f)
```
